```python
import pandas as pd
from pandasai import PandasAI
from pandasai.llm.openai import OpenAI

df = pd.read_csv("/home/shivanksharma4/Downloads/student.csv")
df = df[['name', 'mark', 'gender']]
df
```

```python
template = """You are a chatbot that will interact with a human
Given the following context and a human input,perform the operation given in the input and derive
the insights using the context which would be a dataframe
If the question cannot be answered using the information 
provided answer with "I don't know".

{context}

{chat_history}
Human: {human_input}
Chatbot:"""
```

```python
from langchain import LLMChain
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory
from langchain.chat_models import ChatOpenAI
prompt = PromptTemplate(
    input_variables=["chat_history", "human_input", "context"], template=template
)
memory = ConversationBufferMemory(memory_key="chat_history", input_key="human_input")
chain =LLMChain(
    llm=ChatOpenAI(openai_api_key="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",temperature=0)
    , memory=memory, prompt=prompt
)
```

```python
query="calculate total students of each gender"
result=chain({"context":df , "human_input": query},return_only_outputs=True)
result['text']
```
